```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_points(points, title):
    plt.scatter(points[:, 0], points[:, 1], c='r')
    for i, point in enumerate(points):
        plt.text(point[0], point[1], f'P{i}', fontsize=12, ha='right')
    plt.title(title)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
```

```python
# Original points
points = np.array([[1, 1], [5, 1], [3, 4]], dtype=np.float32)
plt.figure(figsize=(12, 6))
```

```python
# Affine Transformation
affine_matrix = np.array([[1, 0.5, 1], [0.5, 1, -2]], dtype=np.float32)
affine_transformed = cv2.transform(np.array([points]), affine_matrix)[0]
plt.subplot(1, 3, 1)
plot_points(affine_transformed, 'Affine Transformation')
```

```python
# Similarity Transformation
angle = np.radians(30)
scale = 1.2
similarity_matrix = np.array([[scale * np.cos(angle), -scale * np.sin(angle), 2],
                              [scale * np.sin(angle), scale * np.cos(angle), -3]], dtype=np.float32)
similarity_transformed = cv2.transform(np.array([points]), similarity_matrix)[0]
plt.subplot(1, 3, 2)
plot_points(similarity_transformed, 'Similarity Transformation')
```

```python
# Euclidean Transformation
angle = np.radians(45)
euclidean_matrix = np.array([[np.cos(angle), -np.sin(angle), 3],
                             [np.sin(angle), np.cos(angle), -1]], dtype=np.float32)
euclidean_transformed = cv2.transform(np.array([points]), euclidean_matrix)[0]
plt.subplot(1, 3, 3)
plot_points(euclidean_transformed, 'Euclidean Transformation')
```

```python
# Plot original points for reference
plt.figure()
plot_points(points, 'Original Points')

plt.show()
```
