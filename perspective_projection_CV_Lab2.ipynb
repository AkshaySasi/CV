```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def proj_transform(points, H):

  #points to homogeneous coordinates
  n_pts = points.shape[0]
  homogeneous_points = np.hstack((points, np.ones((n_pts, 1))))

  #transformation
  transformed_pts = np.dot(homogeneous_points, H.T)

  #Euclidean ccordinates
  Eucli_points = transformed_pts[:, :2] / transformed_pts[:, 2:]

  return Eucli_points
```

```python

points = np.array([[0, 0],
                   [1, 0],
                   [1, 1],
                   [0, 1]])
```

```python

#projective transformation matrix
H = np.array([[1, 0.5, 0],
              [0.5, 1, 0],
              [0.001,0.001, 1]])
```

```python
transformed_pts = proj_transform(points, H)
transformed_pts
```

```python
plt.scatter(points[:, 0], points[:, 1], color='blue', label='Original Points');
plt.scatter(transformed_pts[:, 0], transformed_pts[:, 1], color='red', label='Transformed Points');
plt.title("Original vs Transformed Points");
plt.legend();
plt.grid(True);
plt.show()
```
